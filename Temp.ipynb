```python
import sympy as sp

t = sp.Symbol("t", real=True, nonnegative=True)
x, y, z = [sp.Function(s)(t) for s in "xyz"]
```

```python
eqs = [sp.Eq(x.diff(t), y ** 2),
       sp.Eq(y.diff(t), x ** 2)]

rhs = [sp.lambdify([x, y], eq.rhs) for eq in eqs]
```

```python
import numpy as np
from scipy.integrate import odeint

def foo(a, t):
    return [f(*a) for f in rhs]

res = odeint(foo, [0.1, 0.2], np.linspace(0, 5))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(res)
plt.legend(["x", "y"])
```
